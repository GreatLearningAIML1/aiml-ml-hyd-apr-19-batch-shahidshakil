```python
#Importing the necessary Libraries 
import pandas as pd
import re
new_list=[]

#Reading the La Liga file and puttting it in a dataframe
liga_data = pd.read_csv('Laliga.csv' , delimiter=',')

print(liga_data.head())

#Replacing the fields which have - to 0
liga_data=liga_data.replace('-',0)

#Converting the data types wherever needed 
liga_data['Seasons']=liga_data['Seasons'].astype(int)
liga_data['Points']=liga_data['Points'].astype(int)
liga_data['GamesPlayed']=liga_data['GamesPlayed'].astype(int)
liga_data['GamesWon']=liga_data['GamesWon'].astype(int)
liga_data['GamesDrawn']=liga_data['GamesDrawn'].astype(int)
liga_data['GamesLost']=liga_data['GamesLost'].astype(int)
liga_data['GoalsFor']=liga_data['GoalsFor'].astype(int)
liga_data['GoalsAgainst']=liga_data['GoalsAgainst'].astype(int)
liga_data['Champion']=liga_data['Champion'].astype(int)
liga_data['Runner-up']=liga_data['Runner-up'].astype(int)
liga_data['Third']=liga_data['Third'].astype(int)
liga_data['Fourth']=liga_data['Fourth'].astype(int)
liga_data['Fifth']=liga_data['Fifth'].astype(int)
liga_data['Fourth']=liga_data['Fourth'].astype(int)
liga_data['Fifth']=liga_data['Fifth'].astype(int)
liga_data['Sixth']=liga_data['Sixth'].astype(int)


##Changing the column Debut to make it optimal for querying

for i in liga_data['Debut']:
    new_list.append(re.sub(r'\W\d\d','',i))
liga_data['Debut_updated']=new_list
liga_data['Debut_updated']=liga_data['Debut_updated'].astype(int)

print(liga_data.dtypes)
```

```python
#Printing all the teams which have started playing between 1930-1980
print("The following teams have debuted between 1930 & 1980")
print('-----------------------------------------------------')
for i in range(len(liga_data)):
    if((liga_data['Debut_updated'][i]>1930) & (liga_data['Debut_updated'][i] < 1980)):
        print(liga_data['Team'][i])
```

```python
#Print the list of teams which came Top 5 in terms of points
print(liga_data.nlargest(5,'Points'))
```

```python
#Writing a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. 
#Using the same function, find the team which has maximum and minimum goal difference###

def Goal_diff_count():
    liga_data['goaldiff'] = liga_data['GoalsFor']-liga_data['GoalsAgainst']
    df1=liga_data[['Team','goaldiff']]
    print(df1)
    print("------------------------------------------------------------")
    print("The team with maximum Goal DIfference is")
    print(df1[df1['goaldiff']==df1['goaldiff'].max()])
    print("------------------------------------------------------------")
    print("The team with least Goal DIfference is")
    print(df1[df1['goaldiff']==df1['goaldiff'].min()])
    
Goal_diff_count()
```

```python
#Create a new column with name “Winning Percent” and append it to the data set (5 points)
#Percentage of Winning = (GamesWon / GamesPlayed)*100
#If there are any numerical error, replace it with 0%
#Print the top 5 teams which has the highest Winning percentage

liga_data['winning_percent']=liga_data['GamesWon']/liga_data['GamesPlayed']*100
liga_data['winning_percent']=liga_data['winning_percent'].fillna(0)
print(liga_data['winning_percent'])
print(liga_data.nlargest(5,'winning_percent'))
```

```python
#Group teams based on their “Best position” and print the sum of their points for all positions
point_data = liga_data['Points'].groupby(liga_data['BestPosition'])
print(point_data.sum())
```
